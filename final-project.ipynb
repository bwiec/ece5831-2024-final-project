## Smart Parking Lot demonstration
To demonstrate this code, we will use a set of images from the CNRPark dataset in `datasets/cnrpark`. In particular, we want to use a set of images taken as though it were a video which is to say we want to have the camera positioned in the same spot. This way, we can designate an x-y coordinate 'zone' for the entrance and exit regions of the image. The command below uses python `glob` package to search teh dataset file structure for all images from camera 6

Because these are not actual videos but still images taken at various times throughout the day, many images do not have any vehicles entering or exiting the parking lot. So we expect to see many frames being processed but not updating the parking lot state. This is demonstrated in the subsequent script - most images are processed and nothing happens. However, when an image is used which *does* have a vehicle in the designated entry/exit zones, matplotlib is used to show the image and RoI as well as the classification. Also, the vehicle state is updated and printed to the console. It will look something like this:

```
Exit vehicle: 
Predicted class: Suzuki Aerio Sedan 2007
predicted_class.item(): 180
{'normal': 26.0, 'compact': 4.0, 'electric_vehicle': 3.0, 'RV': 2.0, 'motorcycle': 4.0, 'emergency': 0.0}
```

Because it printed 'exit vehicle', it means the vehicle was exiting the lot. It then tells you the predicted class and the last line is the printout of the new vehicle state

````python
# From command line, run with: python main.py $(ls datasets/cnrpark/CNR-EXT_FULL_IMAGE_1000x750/FULL_IMAGE_1000x750/*/*/camera6/*)
# From jupyter with %run magic command, need to do use glob to get files into python

# The glob command is searching the dataset file structure for all image files from camera6

from glob import glob
files = glob("datasets/cnrpark/CNR-EXT_FULL_IMAGE_1000x750/FULL_IMAGE_1000x750/*/*/camera6/*")
files_str = " ".join(files)

%run main.py {files_str}
````
